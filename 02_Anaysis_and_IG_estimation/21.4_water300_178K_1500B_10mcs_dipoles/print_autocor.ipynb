```python
import numpy as np
import matplotlib.pyplot as plt 

CB_color_cycle = ['#377eb8', '#ff7f00', '#4daf4a',
                  '#f781bf', '#a65628', '#984ea3',
                  '#999999', '#e41a1c', '#dede00']

plt.rcParams['axes.prop_cycle'] = plt.cycler(color=CB_color_cycle)


plt.rcParams['font.size'] = 18
plt.rcParams['legend.fontsize'] = 12
```

```python
fig, ax = plt.subplots()

ax.set_xscale('log')
ax.set_xlabel("Time (ns)")
ax.set_ylabel("Autocorr (-)")
ax.set_title("Autocorrelation of the different dipoles: LDL")
ax.set_ylim((-0.5,1.1))

Name = "central"
Data = np.loadtxt(f"dipole_{Name}_cor_mean_v3.txt")[1:,:]
Data[:,1]/=Data[0,1]
Data[:,2]/=Data[0,1]
Data[:,0]/=1000000
prevplot = ax.plot(Data[:,0], Data[:,1], label = Name)
ax.fill_between(Data[:,0], Data[:,1]+Data[:,2], Data[:,1]+Data[:,2], lw=0)


Name = "1stshell"
Data = np.loadtxt(f"dipole_{Name}_cor_mean_v3.txt")[1:,:]
Data[:,1]/=Data[0,1]
Data[:,2]/=Data[0,1]
Data[:,0]/=1000000
prevplot = ax.plot(Data[:,0], Data[:,1], label = Name)
ax.fill_between(Data[:,0], Data[:,1]+Data[:,2], Data[:,1]+Data[:,2], lw=0)

# Name = "2ndshell"
# Data = np.loadtxt(f"dipole_{Name}_cor_mean_v3.txt")[1:,:]
# Data[:,1]/=Data[0,1]
# Data[:,2]/=Data[0,1]
# Data[:,0]/=1000000
# prevplot = ax.plot(Data[:,0], Data[:,1], label = Name)
# ax.fill_between(Data[:,0], Data[:,1]+Data[:,2], Data[:,1]+Data[:,2], lw=0)

Name = "out_1stshell"
Data = np.loadtxt(f"dipole_{Name}_cor_mean_v3.txt")[1:,:]
Data[:,1]/=Data[0,1]
Data[:,2]/=Data[0,1]
Data[:,0]/=1000000
prevplot = ax.plot(Data[:,0], Data[:,1], label = Name)
ax.fill_between(Data[:,0], Data[:,1]+Data[:,2], Data[:,1]+Data[:,2], lw=0)

# Name = "out_2ndshell"
# Data = np.loadtxt(f"dipole_{Name}_cor_mean_v3.txt")[1:,:]
# Data[:,1]/=Data[0,1]
# Data[:,2]/=Data[0,1]
# Data[:,0]/=1000000
# prevplot = ax.plot(Data[:,0], Data[:,1], label = Name)
# ax.fill_between(Data[:,0], Data[:,1]+Data[:,2], Data[:,1]+Data[:,2], lw=0)

ax.legend()
ax.grid()
fig.set_size_inches((8,5))
plt.tight_layout()
plt.savefig("autocorr_LDL.pdf", dpi=300)
plt.savefig("autocorr_LDL.png", dpi=300)
plt.show()
```
